```python
import numpy
from matplotlib import pyplot as plt

simulation_num=50
times_taken=500
simulation_time=100000

def random(x,simulation_num=simulation_num):
    elements=[]
    probabilities=[]

    for i in range(simulation_num):
        elements.append(i+1)
        probabilities.append(2**(simulation_num-i))

    probabilities=numpy.array(probabilities)/sum(probabilities)

    return numpy.random.choice(elements, size=x, p=probabilities).tolist()

taken_out=[]

for i in range(times_taken):
    taken_out.append(0)

for k in range(simulation_time):
    balls=[]
    for i in range(simulation_num):
        balls.append(0)

    added=random(5)

    for i in added:
        balls[i-1]+=1
    if k%1000==0:
        print(k+1)
    for i in range(times_taken):
        for j in range(simulation_num,0,-1):
            if balls[j-1]>0:
                biggest=j
                balls[j-1]-=1
                break
        taken_out[i]+=biggest
        added=random(4)
        for j in added:
            balls[j-1]+=1

xArr = []
yArr = []
for i in range(times_taken):
    print('The '+str(i+1)+'th time:',taken_out[i]/simulation_time)
    xArr.append(i+1)
    yArr.append(taken_out[i]/simulation_time)

plt.figure(figsize=(6,3))
plt.xlabel("Survey Round Number",fontsize=8)
plt.ylabel("Expected Number of Stars",fontsize=8)
plt.scatter(xArr,yArr)
plt.show()
```
